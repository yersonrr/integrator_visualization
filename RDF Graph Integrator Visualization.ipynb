#### Usage of the Threshold

A threshold equal to 0.0 does not impose any restriction on the values of similarity; thus the bipartite graph includes all the edges. High thresholds, e.g., 0.8, restrict the values of similarity, resulting in a bipartite graph comprising just a few edges.

```python
Image(filename=path+'threshold.png')
```

Different thresholds on the values of the similarity measure and the impact on a bipartite graph between RDF molecules.

#### Usage of the fusion policy

Using union as a fusion policy only provide a merge between molecules. This fusion policy creates a new molecule and creates a set of (prop, val) pairs where duplicate pairs, and pairs that are syntactically the same,
appear only once.

```python
Image(filename=path+'fusion_union.png')
```

Otherwise, using a URI as a fusion policy does not provide a merge, this fusion policy provides a connection and 
the creation of more triples between the molecules.
